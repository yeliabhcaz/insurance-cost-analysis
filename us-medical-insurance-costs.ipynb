<h1> Comparing Medical Insurance Costs by Region </h1>


<h3>Import Statements</h3>

```python
import pandas as pd
from os import path
from matplotlib import pyplot as plt
```

<h3>Importing Data</h3>

```python
DATA_DIR = '/Users/zachbailey/data-science/medical-insurance-analysis'
insurance_data = pd.read_csv(path.join(DATA_DIR, 'insurance.csv'))
northeast = insurance_data[insurance_data['region'] == 'northeast']
northwest = insurance_data[insurance_data['region'] == 'northwest']
southeast = insurance_data[insurance_data['region'] == 'southeast']
southwest = insurance_data[insurance_data['region'] == 'southwest']
```

<h3>Plotting Cost by Region </h5>
